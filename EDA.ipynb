```python
import numpy as np
import pandas as pd

data = pd.read_csv("match_data.csv",low_memory=False)
data.head()
```

```python
columns_to_replace = ['byes', 'legbyes', 'wides', 'noballs', 'penalty']
data[columns_to_replace] = data[columns_to_replace].fillna(0)
data.isnull().sum()
```

```python
data['TotalByes'] = data['byes'] + data['legbyes']
data.columns
# data.head()
```

```python
data['TotalRuns'] = data['runs_off_bat'] + data['extras'] + data['wides'] + data['penalty'] + data['TotalByes'] + data['noballs']
```

```python
new_data = data.drop(['ball', 'non_striker', 'byes', 'legbyes', 'other_wicket_type', 'other_player_dismissed', 'cricsheet_id'], axis=1)
new_data.columns
```

```python
new_data.isnull().sum()
```

```python
new_data["innings"].unique()
```

```python
new_data["innings"]= new_data["innings"].replace(3, 1)
new_data["innings"]= new_data["innings"].replace(4, 1)
new_data["innings"]= new_data["innings"].replace(5, 2)
new_data["innings"]= new_data["innings"].replace(6, 2)
```

```python
new_data["innings"].unique()
```

```python
# new_data["season"].value_counts()
```

```python
new_data["season"].unique()
```

```python
new_data["season"]= new_data["season"].replace("2020/21", "2020")
new_data["season"]= new_data["season"].replace("2009/10", "2010")
new_data["season"]= new_data["season"].replace("2007/08", "2008")
new_data["season"].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Univariate analysis

gp = new_data.groupby('season')['extras'].sum()
gp.plot(kind='line')
plt.show()
```

```python
sns.countplot(data=new_data, x='innings')
plt.title('Distribution of Innings Across Matches')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=new_data, x='batting_team')
plt.title('Number of Matches Played by Each Batting Team')
plt.xticks(rotation=90)
plt.show()
```

```python
# Bivariate analysis

sns.lineplot(x="season", y="extras",data=new_data, hue="innings", palette="spring")
plt.xticks(rotation=85)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x ='batting_team', y ='TotalRuns', data = new_data)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=new_data, x='batting_team', hue='innings')
plt.title('Number of Matches Played by Each Batting Team per Inning')
plt.xticks(rotation=90)
plt.show()
```
